# Moviev Rating Analysis

## Introduction
This notebook conducts a comprehensive analysis of the IMDB movie dataset, focusing on data cleaning, exploration, and visualization. 
The dataset is first loaded and inspected to understand its structure and identify missing values. To ensure data integrity, missing values in crucial columns such as revenue and metascore are filled with their median and mean, respectively, rather than being dropped, preserving as much data as possible.
 
The analysis proceeds by identifying the top-rated movies and the highest-earning films. Movies are grouped by genre to compute average ratings, providing insights into genre popularity and critical reception. Additionally, directors are ranked based on the average ratings of their films, highlighting the most critically acclaimed filmmakers in the dataset.
 
To explore the relationship between movie ratings and financial performance, a correlation analysis is performed between ratings and revenue. A scatter plot is generated to visually assess whether higher-rated films tend to earn more. The results indicate a weak positive correlation, suggesting that while ratings may have some impact on revenue, other factors such as marketing, budget, and franchise popularity likely play a more significant role.

## Step 3: Clean the Data

### Handling missing values:
Instead of dropping rows with missing values, they are filled to retain as much data as possible. 
Removing rows could lead to a significant reduction in the dataset, affecting the reliability of insights.
The 'Revenue (Millions)' column is highly skewed, with extreme values inflating the mean. To address this, 
the median is used as it provides a more representative central value, reducing the impact of outliers.
The 'Metascore' column follows a more normal distribution, making the mean a suitable replacement 
without distorting the dataset.

```python
cleaned_df = df.copy()
```

```python
# Fill missing values in 'Revenue (Millions)' and 'Metascore' with their mean
cleaned_df['Revenue (Millions)'] = cleaned_df['Revenue (Millions)'].fillna(cleaned_df['Revenue (Millions)'].median())
cleaned_df['Metascore'] = cleaned_df['Metascore'].fillna(cleaned_df['Metascore'].mean())
```

```python
# Trim column names to remove spaces
cleaned_df.columns = cleaned_df.columns.str.strip()
```

## Step 4: Top 10 Highest Rated Movies

```python
top_rated_movies = cleaned_df.sort_values(by='Rating', ascending=False).head(10)
display(top_rated_movies)
```

## Step 5: Gnere Average Rating

```python
df_exploded = cleaned_df.assign(Genre=cleaned_df['Genre'].str.split(',')).explode('Genre')

genre_avg_rating = df_exploded.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
display(genre_avg_rating)
```

```python
plt.figure(figsize=(12, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(genre_avg_rating)))
plt.barh(genre_avg_rating.index, genre_avg_rating.values, color=colors)
plt.xlabel("Average Rating")
plt.ylabel("Genre")
plt.title("Average Rating by Genre")
plt.gca().invert_yaxis()
plt.show()
```

## Step 6: Top 10 Directors Based on Average Rating

```python
top_directors = cleaned_df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(10)
display(top_directors)
```

## Step 7: Find the Top 10 Movies with Highest Revenue

```python
top_revenue_movies = cleaned_df.sort_values(by='Revenue (Millions)', ascending=False).head(10)
display(top_revenue_movies)
```

## Step 8: Check Correlation Between Rating and Revenue

```python
correlation = cleaned_df[['Rating', 'Revenue (Millions)']].corr()
display(correlation)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=cleaned_df['Rating'], y=cleaned_df['Revenue (Millions)'], alpha=0.7)
plt.xlabel("Movie Rating")
plt.ylabel("Revenue (Millions)")
plt.title("Correlation Between Movie Rating and Revenue")
plt.show()
```

### Answer
The correlation matrix shows a weak positive correlation (0.218) between movie ratings and revenue, indicating that higher-rated movies tend to earn slightly more, but the relationship is not strong. The scatter plot supports this, showing that while some highly rated movies have high revenue, many lower-rated films also perform well financially. There is no clear linear trend, suggesting that factors like marketing, budget, and franchise popularity likely have a greater impact on revenue than ratings alone.
